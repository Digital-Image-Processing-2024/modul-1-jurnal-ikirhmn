```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
plt.style.use('ggplot')
```

Code diatas untuk import numpy, cv2 dan matplotlib

```python
image = plt.imread("aang.jpg",0)
plt.imshow(image) 
```

Code diatas berfungsi untuk menampilkan gambar yang sudah diimport sebelumnya.

```python
gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

Code diatas untuk menampilkan gambar yang telah diimport dan diproses menjadi grayscale, sehingga gambar yang akan ditampilkan menjadi berwarna abu.

```python
height, width = gray.shape
print(gray.shape)
print(gray)
```

Code diatas untuk menampilkan array dari citra grayscale

```python
sumGray = np.zeros(256).astype(int)

for i in range (height):
    for j in range (width):
        val = gray [i,j]
        sumGray[val]+=1

sumGray
```

Code diatas untuk membuat sebuah array dengan ukuran 256 dengan seluruh elemnnya bernilai 0

```python
grayVal = [i for i in range(256)]
plt.bar(grayVal,sumGray)
```

Code diatas untuk menampilkan grafik histogram dari intensitas piksel code sebelumnya

```python
plt.bar(grayVal, sumGray/sumGray.sum())
```
